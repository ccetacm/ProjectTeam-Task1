```python
import pandas as pd
from sklearn.model_selection import train_test_split
import joblib
from sklearn.feature_extraction.text import CountVectorizer
import re
import numpy as np
```

```python
data = pd.read_csv('project_train.csv')
data_test=pd.read_csv('project_test.csv')
```

```python
data.head()
```

```python
data.describe
```

```python
def preprocess_data(data):
    data = data.drop('keyword', axis=1)
    data=data.drop('location',axis=1)
    data=data.drop('id',axis=1)
    data['text'] = data['text'].str.strip().str.lower()
    return data
```

```python
data=preprocess_data(data)
```

```python
x=data['text']
y=data['target']
x, x_test, y, y_test = train_test_split(x,y, stratify=y, test_size=0.25, random_state=42)
```

```python
vec = CountVectorizer(stop_words='english')
x = vec.fit_transform(x).toarray()
x_test = vec.transform(x_test).toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(x, y)
```

```python
model.score(x_test, y_test)
```

```python
test_list=[]
test_list=data_test.text.tolist()
```

```python
test_list
```

```python
id_list=[]
id_list=data_test.id.tolist()
```

```python
id_list
```

```python
res_array=[]
for i in test_list:
    s=model.predict(vec.transform([i]))
    res_list.append(s)
```

```python
final_list = []
for sublist in res_list:
    for item in sublist:
        final_list.append(item)
```

```python
list_dict = {'id':id_list, 'target':final_list} 
df = pd.DataFrame(list_dict) 
df.to_csv('submission9.csv', index=False) 
```
